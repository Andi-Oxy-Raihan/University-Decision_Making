```python
import pandas as pd
import numpy as np

data = pd.read_csv("/content/drive/MyDrive/Datasets/(6DM2) Data Tugas.csv")
data
```

```python
#Konversi Nilai Yang Hilang
data.replace(["unknown","?","NaN","nan","NA"], np.nan, inplace=True)
data.isnull().sum()
```

```python
categorical_columns = ['make','bahan-bakar','jumlah-pintu','body-style']
data[categorical_columns] = data[categorical_columns].fillna("unknown")

for col in categorical_columns:
    print("Kategori pada kolom:", col)
    print(data[col].value_counts())
    print("\n")
```

```python
mean_length = data['length'].mean()
mean_width = data['width'].mean()
mean_height = data['height'].mean()

data['length'].fillna(mean_length, inplace=True)
data['width'].fillna(mean_width, inplace=True)
data['height'].fillna(mean_height, inplace=True)
```

```python
data.info()
```

```python
#Min-Max Normalization

min_val = data['wheel-bases'].min()
max_val = data['wheel-bases'].max()

data['wheel-bases-normalized'] = (data['wheel-bases'] - min_val) / (max_val - min_val)
data[['wheel-bases','wheel-bases-normalized']].head()
```

```python
columns = ['length','width','height']

for col in columns:
    data[col + '_zscore'] = (data[col] - data[col].mean()) / data[col].std()

data[['length_zscore','width_zscore','height_zscore']].head()
```
